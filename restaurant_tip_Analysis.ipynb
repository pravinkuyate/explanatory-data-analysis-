```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\pravinkuyate\Downloads\tips.csv.xls")
```

```python
df.rename(columns={"Sex":"Gender"},inplace=True) #replace sex column name with gender
df.head(2) 
```

```python
x=df['day']
y=df['total_bill']
plt.xlabel('day')
plt.ylabel('bill')
plt.bar(x,y,color='yellow',edgecolor='blue',linewidth=1)
plt.show()
```

```python
x=df['total_bill']
plt.hist(x,bins=25,color='pink',edgecolor='red',linestyle='--',alpha=1)
plt.title('tips dataset')
plt.xlabel('frequency')
plt.ylabel('total_bill')
plt.show()
```

```python
df.corr() #corelation of each column
```

```python
x=df['tip']
y=df['total_bill']
plt.scatter(x,y)
plt.title('scatter')
plt.ylabel('total bill')
plt.xlabel('tips')
plt.show()
```

```python
df.value_counts()
```
